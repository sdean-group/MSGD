# The convergence of Loss on Census dataset

```python
import seaborn as sns
import pickle
import matplotlib.pyplot as plt 
import numpy as np
n = 3
eta = 1
file_name = f'./results/census_n{n}_eta{eta}.pickle'
zeta_list = [0, 0.2, 0.5, 0.8, 1]
T = 2000
with open(file_name, 'rb') as file:
    overall_result = pickle.load(file)


fig, axs = plt.subplots(1, 1, figsize = (6, 6))
line_width=2
sns.set_style("white")


for i, zeta in enumerate(zeta_list):
    x = np.log(range(0, T, 1))
    y = np.log(overall_result[i]['total_loss'])
    axs.plot(x, y, label = f'$\zeta$={zeta_list[i]}', linewidth=line_width)
    
x = np.log(range(0, T, 1))
y = np.log(overall_result[0]['total_loss_full'])
axs.plot(x, y, label = f'Full Information', linewidth=line_width)
axs.legend()

```

# The convergence of iterate on Census dataset

```python
import seaborn as sns
import pickle
import matplotlib.pyplot as plt 
import numpy as np
n = 3
eta = 1
file_name = f'./results/census_n{n}_eta{eta}.pickle'
zeta_list = [0, 0.2, 0.5, 0.8, 1]
T = 2000
with open(file_name, 'rb') as file:
    overall_result = pickle.load(file)


fig, axs = plt.subplots(1, 1, figsize = (6, 6))
line_width=2
sns.set_style("white")


for i, zeta in enumerate(zeta_list):
    x = range(0, T-1, 1)
    y = np.linalg.norm(np.diff(overall_result[i]['Theta'], axis = -1), axis =1)
    y = np.mean(np.cumsum(y, axis =1), axis = 0)
    axs.plot(x, y, label = f'$\zeta$={zeta_list[i]}', linewidth=line_width)
x = range(0, T-1, 1)
y = np.linalg.norm(np.diff(overall_result[i]['Theta_full'], axis = -1), axis =1)
y = np.mean(np.cumsum(y, axis =1), axis = 0)/3
axs.plot(x, y, label = f'Full Information', linewidth=line_width)
axs.legend()
```
